```python
data = {
    "Name": ["Ali", "dua", "Ahmed", None, "Sara","dua","dua"],
    "Marks": [85, 90, 78, 85, None,70,80],
    "City": ["Lahore", "Karachi", "Islamabad", None, "Karachi","fsd","fsd"]
}

df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd
data1={
    "ID":[1,2,3,4,5,6,10],
     "name":["dua","malika","uswa","wajiha","soha","tayyaba","hamna"]
}
df1=pd.DataFrame(data1)
print(df1)
```

```python
data2 = {
    "ID":[1,2,9,4,5,8,7],
    "Marks":[95,90,70,80,45,67,89]
}
df2=pd.DataFrame(data2)
print(df2)
```

```python
df=pd.merge(df1,df2,on="ID")
print(df) # merging jo ha i wo sirf tb use ho gi jb hamary pass same ho koi cheez means id
```

```python
pd.merge(df1,df2,on="ID",how="inner")# inner jo hai wo sirf commom wali value write kery ga 
```

```python
pd.merge(df1,df2,on="ID",how="outer") #outer jo hai wo sara likhy ga commom wali ko aik dafa likhy ga
```

```python
pd.merge(df1,df2,on="ID",how="left")#df1 left wala data likhy ga
```

```python
pd.merge(df1,df2,on="ID",how="right")
```

```python
pd.concat([df1,df2],axis=1)#in the form of column concat kery ga
```

```python
pd.concat([df1,df2],axis=0)# row wise concat ker ga
```

```python
ages = [12, 18, 25, 30, 45, 60]
df = pd.DataFrame({"Age": ages})
print(df)
```

```python
df["Age_Group"] = pd.cut(df["Age"],
                         bins=[0,18,30,50,70],
                         labels=["Teen","Young","Adult","Senior"])
print(df)
```

```python
df["Age_Group"] = pd.qcut(df["Age"], q=3)
print(df)
```
